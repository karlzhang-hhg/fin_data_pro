```python
import re
import copy
import numpy as np

def Append_Comp_Info(df_raw, df_target):
    """Preprocess a raw dataframe from company information and append the results to target dataframe."""
    row_num = df_target.shape[0]
    print("The raw data has number of rows %s \n The target dataframe have number of rows %s" % (df_raw.shape[0], row_num))
    # This set is for the suffix of shareholder names. Those suffix are preceded by comma which make the parsing of shareholders
    # problematic.
    suffix_set = set(['L\.P\.', 'Inc', 'INC', 'S\.L', 'S\.A\.', 'LLC', 'Ltd', 'Llc', 'Lp', 'L\.P', 'Sl', 'Obras', 'Ltda', 'Lda',
                      'Cuniff', 'Cunniff', 'Mayo', 'Redito', 'S\sL', 'S\.A\.B\.', 'C\.V\.', 'S\s', 'V\s', 'com\s',
                      'National\sAssociation', 'Importacao', 'Brasil\sLtda', 'Participacao', 'Credito', 'Financiamento', 'Comercio',
                      'Sociedade'])
    suffix_cnpj_set = set([u'01545826000107', u'33611500000119', u'10808462000193', u'04065033000170', u'03847461000192', 
                           u'09191251000175', u'14578002000177', u'02387241000160', u'07207197000183', u'33014556000196', 
                           u'08322396000103', u'22677520000176', u'33022203000138', u'11284204000118', u'08902291000115', 
                           u'13407617000178', u'61409892000173', u'12944170000103', u'02221358000170', u'08764621000153', 
                           u'02619221000178', u'02191601000154', u'03870455000580', u'87376109000106', u'29978814000187', 
                           u'16614075000100', u'58514928000174', u'18499616000114', u'02440482000126', u'04437534000130', 
                           u'60398369000126', u'40281347000174', u'02102498000129', u'34075739000184', u'06945520000153', 
                           u'01083200000118', u'03002339000115', u'10807972000146', u'02558157000162', u'71550388000142', 
                           u'54517628000198', u'30937874000130', u'72908353000103', u'33592510000154', u'09149503000106', 
                           u'63310411000101', u'76483817000120', u'03537249000129', u'04221023000187', u'89545511000100', 
                           u'33931494000187', u'02800026000140', u'50290329000102', u'60872504000123', u'00001180000126', 
                           u'14807945000124', u'90400888000142', u'09080006000190', u'06977751000149', u'43470988000165', 
                           u'75315333000109', u'71371686000175', u'14868712000131', u'03983431000103', u'08446503000105', 
                           u'92754738000162', u'50306471000109', u'03220438000173', u'07857850000150', u'02917443000177', 
                           u'09346601000125', u'00950724000104', u'42564807000105', u'33170085000105', u'02009924000184', 
                           u'07043628000113', u'95424321000120', u'07021544000189', u'11055041000100', u'05648257000178', 
                           u'06977745000191', u'47067525000108', u'90852914000173', u'11395604000109', u'09509535000167', 
                           u'05692190000179', u'84685106000166', u'42123091000100', u'04834395000189', u'01961898000127', 
                           u'02380162000128', u'03502099000118', u'33040981000150', u'07689002000189', u'61585865000151', 
                           u'12708358000152', u'09060964000108', u'08610386000165', u'05808792000149', u'40432544000147', 
                           u'72712201000122', u'07779299000173', u'09305994000129', u'02998301000181', u'08078847000109', 
                           u'05707616000110', u'09462140000156', u'66703554000163', u'11744699000110', u'02365069000144', 
                           u'03112879000151', u'02919555000167', u'03065552000176', u'62255682000130', u'08505736000123', 
                           u'03853896000140'])
    CNPJ_set = set()
    for idx, row in df_raw.iterrows():
        # Each row have following value:['Index_Num','Company','Key_Executives','Employee_Number','Employee_Info_Year',
        #                                'Financial_Auditors','Main_Products','Phone','Email','Business_Description','Shareholder_String',
        #                                'Revenue_2017','Revenue_2016','Revenue_2015','Revenue_2014','Revenue_2013','Revenue_2012',
        #                                'Revenue_Estimated_Klooks','Industry_List','NAICS_Num_List','Status','CNPJ',
        #                                'Audited','Consolidated','Shareholder_Num','Shareholder_Index','Shareholder','Equity']
        if np.isnan(row['Num']): # if this row is empty, skip
            print("The row %s is invalid (%s)" % (idx, row['Num']))
            continue
        row_temp = []
        # Direct copy columns
        row_temp.extend(row[['Num','Company','Key Executives','Number of Employees','Employee Reference date','Financial Auditors',
                             'Main Products','Phone','Email','City','State/County','Address','Business Description/Products',
                             'Shareholders']])
        # Total operating revenue
        if re.match(r'2017',row['Fiscal Year']):
            row_temp.extend([row['Total operating revenue'], np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        elif re.match(r'2016',row['Fiscal Year']):
            row_temp.extend([np.nan, row['Total operating revenue'], np.nan, np.nan, np.nan, np.nan, np.nan])
        elif re.match(r'2015',row['Fiscal Year']):
            row_temp.extend([np.nan, np.nan, row['Total operating revenue'], np.nan, np.nan, np.nan, np.nan])
        elif re.match(r'2014',row['Fiscal Year']):
            row_temp.extend([np.nan, np.nan, np.nan, row['Total operating revenue'], np.nan, np.nan, np.nan])
        elif re.match(r'2013',row['Fiscal Year']):
            row_temp.extend([np.nan, np.nan, np.nan, np.nan, row['Total operating revenue'], np.nan, np.nan])
        elif re.match(r'2012',row['Fiscal Year']):
            row_temp.extend([np.nan, np.nan, np.nan, np.nan, np.nan, row['Total operating revenue'], np.nan])
        else:
            row_temp.extend([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, row['Total operating revenue']])
        # Industy and NAICS numbers
        row_temp.append(';'.join(re.findall(r'([^();]+)\([0-9]+\)',row['Industry'])))# For industry name
        # Cannot use comma, because this would mix with comma from csv
        row_temp.append(';'.join(re.findall(r'\(([0-9]+)\)',row['Industry'])))# For NAICS number
        # Operating status
        row_temp.append(int(bool(re.match(r'operational',row['Operational Status'].lower()))))
        # CNPJ
        CNPJ = ''.join(re.findall(r'CNPJ\s*:\s*([0-9]{8})/([0-9]{4})-([0-9]{2})',row['National Fiscal IDs'])[0])
        if CNPJ in CNPJ_set: # Check if any CNPJ is duplicated
            print("The index: %s; The company name: %s; The CNPJ: %s" % (row["Num"], row["Company"], CNPJ))
        else:
            CNPJ_set.add(CNPJ)
        row_temp.append(CNPJ)
        # Audited or not
        row_temp.append(int(bool(re.match(r'Y',row['Audited']))))
        # Consolidated or not
        row_temp.append(int(bool(re.match(r'Consolidated',row['Consolidated']))))
        # Separate shareholders in shareholder list
        try:
            shareholder_list = row['Shareholders'].split(',')
        except AttributeError as e:
            shareholder_list = []
            # print("Index: %s; The error is %s" % (idx,e))
        if shareholder_list:
            row_temp.append(len(shareholder_list))
            for sh_idx, sh in enumerate(shareholder_list):
                row_temp1 = copy.copy(row_temp)# Have to deep copy
                row_temp.append(sh_idx)
                sh = sh.strip()
                equity = re.findall(r'\(([0-9\.]+)%\)', sh, re.U)
                if equity: # Exist equity information
                    sh_name = re.findall(r'(.+)\([0-9\.]+%\)', sh, re.U)
                    if sh_name:
                        if CNPJ in suffix_cnpj_set:
                            skip_flag = False
                            for ele in suffix_set:
                                if re.match(r'^'+ele+'.*', sh_name[0].strip()):
                                    df_target.at[row_num-1,'Shareholder'] += ' ' + sh_name[0].strip()
                                    df_target.at[row_num-1,'Equity'] = float(equity[0])
                                    skip_flag = True
                                    break
                            if skip_flag:# If shareholder has a comma in name, skip this iteration
                                row_temp = row_temp1
                                continue                                
                        row_temp.append(sh_name[0].strip())
                    else: # Some times they mistakenly add comma between name and equity
                        row_temp.append(u'')
                        print('Index: %s; Shareholders: %s' % (row['Num'], row['Shareholders']))
                    row_temp.append(float(equity[0]))
                else: # Only has Shareholder information
                    row_temp.extend([sh, np.nan])
                df_target.loc[row_num] = row_temp
                row_num += 1
                row_temp = row_temp1                    
        else:
            row_temp.extend([np.nan,np.nan,'',np.nan])
            df_target.loc[row_num] = row_temp
            row_num += 1
            
    print("Current number of rows of the target dataframe %s; The number of unique CNPJ is %s" % (df_target.shape[0], len(CNPJ_set)))
```

```python
import numpy as np
import re

def Append_Fina_Info(df_raw, df_target):
    """Preprocess a raw dataframe from financial information and append the results to target dataframe."""
    row_num = df_target.shape[0]
    print("The raw data has number of rows %s \n The target dataframe have number of rows %s" % (df_raw.shape[0], row_num))
    CNPJ_set = set()
    # To automate the 
    fina_column_name_level0_list = ['Cash and Cash Equivalents', 'Current assets', 
                                  'Current liabilities', 'Current loans and borrowings', 
                                  'Depreciation, amortization and impairment charges', 'EBITDA', 
                                  'Employee benefit expense', 'Inventories', 'Net Cash', 'Net Debt', 
                                  'Net Profit (Loss) for the Period', 'Net sales revenue', 'Non-current assets', 
                                  'Non-current liabilities', 'Operating profit (EBIT)', 'Profit before income tax', 
                                  'Raw materials and consumables used', 'Short term financial assets', 
                                  'Total assets', 'Total equity', 'Total liabilities', 'Total operating revenue', 
                                  'Trade and other payables', 'Trade and other receivables']
    for idx, row in df_raw.iterrows():
        # Each row have following value:['Index_Num','Company','Industry_List','NAICS_Num_List','CNPJ',
        #                                'Total_assets'*6,'Current assets'*6,'Inventories'*6,'Trade_and_other_receivables'*6,
        #                                'Cash_and_Cash_Equivalents'*6,'Short_term_financial_assets'*6,'Non-current_assets'*6,
        #                                'Total_equity'*6,'Total liabilities'*6,'Non-current_liabilities'*6,
        #                                'Current_liabilities'*6,'Current_loans_and_borrowings'*6,'Trade_and_other_payables'*6
        #                                'Net_Cash'*6,'Net_Debt'*6,'Raw_materials_and_consumables_used'*6,
        #                                'Employee benefit expense'*6,'EBITDA'*6,'Net_sales_revenue'*6,
        #                                'Depreciation_amortization_and_impairment_charges'*6,'Operating_profit(EBIT)'*6,
        #                                'Profit_before_income_tax'*6,'Net_Profit(Loss)_for_the_Period'*6,
        #                                'Total_operating_revenue'*6,'Audited_Consolidated'*6]
        if np.isnan(row['Num'][0]): # if this row is empty, skip; notice that this is multiindex so that I have to use [0]
            print("The row %s is invalid (%s)" % (idx, row['Num']))
            continue
        row_temp = []
        # Direct copy columns
        row_temp.extend(list(row['Num'])+list(row['Company']))
        # Industy and NAICS numbers
        row_temp.append(';'.join(re.findall(r'([^();]+)\([0-9]+\)',row['Industry'][0])))# For industry name
        # Cannot use comma, because this would mix with comma from csv
        row_temp.append(';'.join(re.findall(r'\(([0-9]+)\)',row['Industry'][0])))# For NAICS number
        # CNPJ
        CNPJ = ''.join(re.findall(r'CNPJ\s*:\s*([0-9]{8})/([0-9]{4})-([0-9]{2})',row['National Fiscal IDs'][0])[0])
        if CNPJ in CNPJ_set: # Check if any CNPJ is duplicated
            print("The index: %s; The company name: %s; The CNPJ: %s" % (row["Num"], row["Company"], CNPJ))
        else:
            CNPJ_set.add(CNPJ)
        row_temp.append(CNPJ)
        # All the 5 year financial information columns
        for col_name in fina_column_name_level0_list:
            row_temp.extend(list(row[col_name]))
            #not use the following line of code because we cannot ensure the order in that way
        #row_temp.extend(list(row[fina_column_name_level0_list]))
        # 'Audited, Consolidated'
        AC_res = list(row['Audited, Consolidated'])
        Audited = []
        Consolidated = []
        for ele in AC_res:
            # Search for 'Audited' or 'Non-Audited'
            try:
                audited_re = re.findall(r'Non-Audited',ele)
                if audited_re:
                    Audited.append(0)
                else:
                    Audited.append(1)
            except TypeError as e:
                Audited.append(np.nan)
            # Search for 'Consolidated' or 'Individual'
            try:
                audited_re = re.findall(r'Individual',ele)
                if audited_re:
                    Consolidated.append(0)
                else:
                    Consolidated.append(1)
            except TypeError as e:
                Consolidated.append(np.nan)
        row_temp.extend(Audited)
        row_temp.extend(Consolidated)
        df_target.loc[row_num] = row_temp
        row_num += 1
            
    print("\nCurrent number of rows of the target dataframe %s; The number of unique CNPJ is %s" % (df_target.shape[0], len(CNPJ_set)))
```
